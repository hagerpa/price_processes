# Black-Scholes and different initial laws

```python
model = BlackScholesModel(volatility=0.2, drift=0.1)
n_samples = 1000
n_steps = 100
T = 1

X = model.sample_paths(T, n_steps, n_samples, start_law = ('dirac', 100))
t_space = np.linspace(0, T, n_steps)
plt.plot(t_space, X[0])
```

```python
n_samples = 3

X = model.sample_paths(T, n_steps, n_samples, start_law = ('diracs', [90,100,110]))
t_space = np.linspace(0, T, n_steps)
plt.plot(t_space, X[0])
plt.plot(t_space, X[1])
plt.plot(t_space, X[2])
```

```python
n_samples = 10

X = model.sample_paths(T, n_steps, n_samples, start_law = ('normal', 100, 20))
t_space = np.linspace(0, T, n_steps)
plt.plot(t_space, X[:,:,0].T);
```

# Bennedsen Fractional Electricity Price

```python
model = BennedsenElectricityModel(volatility=0.2, mean=np.log(100), alpha=-0.4, mean_reversion=0.02)
n_samples = 1000
n_steps = 100
T = 1

X = model.sample_paths(T, n_steps, n_samples, start_law = ('stationary'))
t_space = np.linspace(0, T, n_steps)
plt.plot(t_space, X[0]);
```

# Two Factor Jump Diffusion

```python
model = TwoFactorJumpDiffusion(jump_std=1.0)
n_samples = 1000
n_steps = 100
T = 1

X = model.sample_paths(T, n_steps, n_samples, start_law = ('dirac', (model.mean + model.jump_mean)/2))
t_space = np.linspace(0, T, n_steps)
plt.plot(t_space, X[0]);
```
